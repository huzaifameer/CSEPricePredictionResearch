```python
import pandas as pd

df = pd.read_excel('./data/sinhala_biz_news.xlsx').head(4352)
df['date'] = pd.to_datetime(df['date']).dt.date
```

```python
sentiment_map = {
    'Positive': 1,
    'Neutral': 0,
    'Negative': -1
}

company_columns = [
    'WATA.N0000','AGPL.N0000','HAPU.N0000','MCPL.N0000',
    'KOTA.N0000','BFL.N0000','RWSL.N0000','DIPP.N0000',
    'MGT.N0000','HEXP.N0000'
]

for col in company_columns:
    df[col] = df[col].map(sentiment_map)
```

```python
daily_sentiment = df.groupby('date')[company_columns].mean().reset_index()

daily_sentiment.rename(
    columns={c: f"{c}_sentiment_score" for c in company_columns},
    inplace=True
)

daily_sentiment.to_excel("daily_company_sentiment_scores.xlsx", index=False)
daily_sentiment.head()
```
